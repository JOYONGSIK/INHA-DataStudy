### index/column 설정 

```python
df1 = pd.read_csv('iphone.csv', index_col = 0)
df1
```

```python
df1.rename(columns = {'출시일' : '출시'})
```

```python
df1.index.name = '아이폰' # 인덱스 이름을 넣어줌
df1
```

```python
df1.set_index('출시일') # inplace 를 걸어주지 않았음.
```

```python
df1['핸드폰 기종'] = df1.index
df1
```

```python
df1.set_index('출시일', inplace=True)
df1
```
